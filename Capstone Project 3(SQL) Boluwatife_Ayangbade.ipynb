```python
#importing the SQLite3 module for working with SQLite databases and the Pandas library for data manipulation and analysis.

import sqlite3
import pandas as pd
```

```python
# Load your dataset
df = pd.read_csv('cleaned_census_data.csv')
```

```python
# Connect to a database (will create if not exists)
conn = sqlite3.connect('census_data.db')
cursor = conn.cursor()

# Write records stored in a DataFrame to a SQL database
df.to_sql('census', conn, if_exists='replace', index=False)
```

### Number of males and females are within the age range 70 – 100.

```python
# Define the SQL query to count individuals by gender within the age range of 70 to 100
query1 = """
SELECT Gender, COUNT(*) AS Count
FROM census
WHERE Age BETWEEN 70 AND 100
GROUP BY Gender;
"""

# Execute the SQL query and store the result in a Pandas DataFrame
df_query1 = pd.read_sql_query(query1, conn)

# Display the DataFrame containing the query result
df_query1
```

### Number of students that live in Sagittarius Avenue.

```python
# Define the SQL query to count the number of students living on Sagittarius Avenue
query2 = """
SELECT COUNT(*) AS StudentCount
FROM census
WHERE Street = 'Sagittarius Avenue' AND Occupation = 'Student';
"""

# Execute the SQL query and store the result in a Pandas DataFrame
df_query2 = pd.read_sql_query(query2, conn)

# Display the DataFrame containing the query result
df_query2
```

### The population of every religion in the UK.

```python
# Define the SQL query to count the population of each religion
query3 = """
SELECT Religion, COUNT(*) AS Population
FROM census
GROUP BY Religion;
"""

# Execute the SQL query and store the result in a Pandas DataFrame
df_query3 = pd.read_sql_query(query3, conn)

# Display the DataFrame containing the query result
df_query3
```

### Removing the ‘Jewish’ from the total population

```python
# Define the SQL query to count the population of each religion excluding 'Jewish'
query4 = """
SELECT Religion, COUNT(*) AS Population
FROM census
WHERE Religion != 'Jewish'
GROUP BY Religion;
"""

# Execute the SQL query and store the result in a Pandas DataFrame
df_query4 = pd.read_sql_query(query4, conn)

# Display the DataFrame containing the query result
df_query4
```

```python
# Close the connection to the SQLite database
conn.close()
```
